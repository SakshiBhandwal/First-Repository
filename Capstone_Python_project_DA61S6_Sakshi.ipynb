```python
df = pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
df  

# Reading the dataset.
```

```python
import json
```

```python
data = df.copy()
```

```python
# storing the json format columns in a list

dict_columns = ['genres','keywords','production_companies','production_countries','spoken_languages']
```

```python
for column in dict_columns:
    data[column] = data[column].apply(lambda x:[item['name'] for item in json.loads(x)])
    
# the lambda funcrion takes input as x and the json.loads(x) converts the json string into python dictionary
# the list comprehension extracts the value associated with the kay name [item['name']] for item
```

```python
data
```

```python
list_columns = ['genres','keywords','production_companies','production_countries','spoken_languages']

# storing the list format columns in the above list.
```

```python
for column in list_columns:
    data[column] = data[column].str[0]
```

```python
data
```

```python
# dealing with the null values.

data.isnull().sum()
```

```python
# Title and genres of first 50 columns is displayed

data[['title','genres']].head(50)
```

# Task 2

## Handling the Null Values for Data Processing.

```python
# dropped the homepage column as it contains only URL of the movie which is not necessary for the analysis

data = data.drop(['homepage'],axis=1)
```

```python
# calculating the mode for keywords column

keywords_mode = data['keywords'].mode()[0]
keywords_mode
```

```python
# filling the null values of keywords with the mode value.

data['keywords'].fillna(keywords_mode,inplace=True)
```

```python
# calculating the mode for genres column.

genres_mode = data['genres'].mode()[0]
genres_mode
```

```python
# filling the null values of genre with the mode value.

data['genres'].fillna(genres_mode,inplace=True)
```

```python
# calculating the mode for Production company column.

Production_company_mode = data['production_companies'].mode()[0]
Production_company_mode
```

```python
# filling the null values of production_companies with the mode value.

data['production_companies'].fillna(Production_company_mode,inplace=True)
```

```python
# calculating the mode for Production countries column.

production_countries_mode = data['production_countries'].mode()[0]
production_countries_mode
```

```python
data['production_countries'].fillna(production_countries_mode,inplace=True)
```

```python
# calculating the mode for runtime column.

runtime_mode = data['runtime'].mode()[0]
runtime_mode
```

```python
# filling the null values of runtime column with the mode value.

data['runtime'].fillna(runtime_mode,inplace=True)
```

```python
release_date_mode = data['release_date'].mode()[0]
release_date_mode
```

```python
data['release_date'].fillna(release_date_mode,inplace=True)
```

```python
spoken_languages_mode = data['spoken_languages'].mode()[0]
spoken_languages_mode
```

```python
data['spoken_languages'].fillna(spoken_languages_mode,inplace=True)
```

```python
# overview and tagline can't be edited as mode or any assumption value therefore dropping their null values.

data = data.dropna()
```

# Task 7

```python
# calculating the correlation between budget and popularity

correlation = data['budget'].corr(data['popularity'])
correlation.round(2)

# correlation is positive, hence there is a positive relationship between the budget and popularity.
```

```python
# calculating the correlation between budget and revenue
corr = data['budget'].corr(data['revenue'])
corr.round(2)

# correlation is near to zero hence, there's no systematic relatioship between revenue & budget.
```

```python
# correlation between  revenue and popularity.
cor = data['popularity'].corr(data['revenue'])
cor.round(2)
# There is a positive relationship between the revenue and popularity.
```

```python
print('Budget & Popularity ',correlation.round(2))
print('Budget & Revenue    ',corr.round(2))
print('Popularity & Revenue',cor.round(2))
```
